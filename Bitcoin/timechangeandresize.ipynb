```python
import pandas as pd
import datetime
```

```python
data = pd.read_csv('coinbaseUSD_1-min_data_2014-12-01_to_2018-03-27.csv')
data.head()
```

```python
updated_time = [datetime.datetime.fromtimestamp(int(timestamp)).strftime('%Y-%m-%d') for timestamp in data.Timestamp]
```

```python
data['Date'] = updated_time
```

```python
data = data.drop(columns='Timestamp')
```

```python
data.head(4100)
```

```python
grouped_data = data.groupby('Date').last()
```

```python
grouped_data
```

```python
grouped_data.to_csv('coinbase_bitcoin.csv')
```
