```python
import tweepy, datetime, time
import pandas as pd
import numpy
import sys

from config import consumer_key, consumer_secret, access_token, access_token_secret

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
Time=[]
Text=[]
RTCount=[]
Followers=[]
counter = 1

for x in range(1,2000):

    
    tweets = api.user_timeline("@realDonaldTrump", page=x)

    for tweet in tweets:

        if 1417305600<int(time.strftime('%s', time.strptime(tweet["created_at"],'%a %b %d %H:%M:%S +0000 %Y')))<1522108800:
            Time.append(tweet["created_at"])
            Text.append(tweet["text"])
            RTCount.append(tweet["retweet_count"])
            Followers.append(tweet["user"]["followers_count"])
            print(f'Tweet {counter}: {tweet["created_at"]}')
            
            counter = counter + 1
    
```

```python
Trump=pd.DataFrame(data={"Time":Time,
                    "Text":Text,
                    "RTCount":RTCount,
                    "Followers":Followers})
```
