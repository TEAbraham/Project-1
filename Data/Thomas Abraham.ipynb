```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import pandas as pd
import numpy as np

from datetime import datetime
```

```python
trump_tweet = pd.read_csv('cleaned_merged.csv')
nasdaq = pd.read_csv('Nasdaq.csv')
dow_jones = pd.read_csv('DJI.csv')
sp500 = pd.read_csv('SP_500.csv')
bitcoin = pd.read_csv('coinbase_bitcoin.csv')
```

```python
x = ['2013','2014','2015','2016','2017','2018']

sns.set(rc={'figure.figsize':(20,10)})
g=sns.barplot(x=trump_tweet.Date, y=trump_tweet['# of Text'])

g.xaxis.set_major_locator(ticker.MultipleLocator(359))
g.xaxis.set_major_formatter(ticker.FixedFormatter(x))

plt.title(f"Trump's Frequency Over Time")
plt.ylabel("Tweet Polarity")
plt.ylim(0,200)
plt.show()
 
```

```python
x = ['2013','2014','2015','2016','2017','2018']

sns.set(rc={'figure.figsize':(20,10)})
g=sns.barplot(x=trump_tweet.Date, y=trump_tweet.Avg_Followers)

g.xaxis.set_major_locator(ticker.MultipleLocator(359))
g.xaxis.set_major_formatter(ticker.FixedFormatter(x))

plt.title(f"Trump's Followers Over Time")
plt.ylabel("Tweet Polarity")
plt.ylim(0,50000000)
plt.show()
```
